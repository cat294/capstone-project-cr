Introduction/ Business Problem

This research is meant to solve a problems faced by employees who are being transferred to the city of Chennai in large numbers due to relocation of various IT companies there. In the study we leverage Foursquare location data to find patterns and clusters of restaurants to make the lives easier of those employees that do infact agree to being transferred to the city. This study focusses to find out if there are ample number of restaurants in the area so that the employees can live in those areas that have abundant restaurants. As setting up offices in Chennai is very economical for companies, this seems to be a desired solution for companies wishing to make large budgetary savings on cutting operational costs by moving staff to the city of Chennai. 


Data and Methodology

In this study I will be using Foursquare location data to solve the problem of finding neighborhoods in Chennai that have a large number of restaurants/food stalls/cafes. 
Firstly, the number of neighborhoods and their respective coordinates need to be retrieved from wikipedia, so that Foursquare can find nearby venues. Using this data, Foursquare should search for nearby venues and get their categories.
These venues are then clustered using k-means. The cluster in which eateries are of the highest frequency will be the set of neighborhoods we are looking for. All of these neighborhoods are suitable for the employees to move in.
This problem can also be easily extended to fit other requests, such as finding the neighborhoods with low flat prices, neighborhoods with a wide variety of shops and malls, neighborhoods with good access to public transport systems etc.
The target audience here are employees who are moving to a new city and require some knowledge about the neighborhoods beforehand so that they can decide the place they want to live in.
